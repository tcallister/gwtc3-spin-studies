```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('plotting.mplstyle')
import numpy as np
```

```python
# Load sampleDict
sampleDict = np.load("../code/input/sampleDict_FAR_1_in_1_yr.pickle", allow_pickle=True)
```

```python
# Events we want to plot
events = ['S190517h', 'S191103a', 'GW150914']
labels = ['GW190517\\_055101', 'GW191103\\_012549', 'GW150914']

# Make figure
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 4.5))

for ax,event,label in zip(axes, events, labels): 
    
    ax.set_rasterization_zorder(2)
    
    chi1s = sampleDict[event]['a1']
    chi2s = sampleDict[event]['a2']
    
    # downsample
    idxs = np.random.choice(len(chi1s), size=5000)
    
    ax.scatter(chi1s[idxs], chi2s[idxs], s=7, zorder=1)
    
    ax.text(0.05, 0.91, label, fontsize=14, bbox=dict(facecolor='white', alpha=0.9))
    
    ax.set_xlabel(r'$\chi_1$')
    ax.set_ylabel(r'$\chi_2$')
    ax.set_xlim((0,1))
    ax.set_ylim((0,1))

plt.subplots_adjust(wspace=0.3)
plt.savefig('figure_09.pdf', bbox_inches='tight', dpi=200)
plt.show()
```
